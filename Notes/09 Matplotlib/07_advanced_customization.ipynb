```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Annotations

Annotations help highlight important points or add explanatory text to your plots.


```python
# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Find maximum point
max_idx = np.argmax(y)
max_x = x[max_idx]
max_y = y[max_idx]

# Create plot with annotation
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, linewidth=2)
ax.plot(max_x, max_y, 'ro', markersize=10)

# Add annotation
ax.annotate('Maximum', xy=(max_x, max_y), xytext=(max_x + 1, max_y + 0.3),
           arrowprops=dict(arrowstyle='->', color='red', lw=2),
           fontsize=12, fontweight='bold', color='red')

ax.set_xlabel('X')
ax.set_ylabel('sin(X)')
ax.set_title('Plot with Annotation')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Multiple Y-Axes

Sometimes you need to plot data with different scales on the same plot.


```python
# Create data with different scales
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # Range: -1 to 1
y2 = x * 10     # Range: 0 to 100

fig, ax1 = plt.subplots(figsize=(10, 6))

# First y-axis
color1 = 'steelblue'
ax1.set_xlabel('X values', fontsize=12)
ax1.set_ylabel('sin(X)', color=color1, fontsize=12)
line1 = ax1.plot(x, y1, color=color1, linewidth=2, label='sin(X)')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, alpha=0.3)

# Second y-axis
ax2 = ax1.twinx()
color2 = 'coral'
ax2.set_ylabel('10*X', color=color2, fontsize=12)
line2 = ax2.plot(x, y2, color=color2, linewidth=2, label='10*X')
ax2.tick_params(axis='y', labelcolor=color2)

# Title
plt.title('Multiple Y-Axes Example', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Logarithmic Scales

Logarithmic scales are useful when data spans several orders of magnitude.


```python
# Create data spanning multiple orders of magnitude
x = np.linspace(1, 1000, 100)
y = np.exp(x / 100)  # Exponential growth

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Linear scale
axes[0].plot(x, y, linewidth=2)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('Linear Scale')
axes[0].grid(True, alpha=0.3)

# Logarithmic scale
axes[1].plot(x, y, linewidth=2)
axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_xlabel('X (log scale)')
axes[1].set_ylabel('Y (log scale)')
axes[1].set_title('Logarithmic Scale')
axes[1].grid(True, alpha=0.3, which='both')

plt.tight_layout()
plt.show()
```
