## What is Matplotlib?

**Matplotlib** is a comprehensive library for creating static, animated, and interactive visualizations in Python. It provides:

- **Figure**: The entire window/page where plots are drawn
- **Axes**: The actual plot area (where data is visualized)
- **Plot types**: Line plots, bar charts, scatter plots, histograms, and many more
- **Customization**: Full control over colors, labels, titles, legends, and styling
- **Integration**: Works seamlessly with NumPy arrays and Pandas DataFrames

**Why Matplotlib for Data Engineering?**
- **Data visualization**: Essential for exploring and understanding data
- **Reporting**: Create professional charts for reports and dashboards
- **Data quality checks**: Visualize data distributions, outliers, and patterns
- **ETL monitoring**: Plot data pipeline metrics and performance
- **Integration**: Works directly with Pandas (which you already know!)
- **Industry standard**: Most widely used plotting library in Python
- **Flexibility**: Can create publication-quality figures

## Understanding Matplotlib Architecture

Matplotlib has a hierarchical structure:

1. **Figure**: The top-level container (the entire window/page)
   - Can contain multiple plots (subplots)
   - Controls overall size and layout

2. **Axes**: The actual plotting area (where your data is drawn)
   - Each Axes has x-axis and y-axis
   - You can have multiple Axes in one Figure

3. **Plot elements**: Lines, bars, markers, text, legends, etc.

**Two approaches:**
- **pyplot (plt)**: Simple, MATLAB-like interface (easier for beginners)
- **Object-oriented (OO)**: More control, better for complex plots

We'll start with pyplot as it's simpler and more intuitive.


## Your First Plot: Line Plot

A line plot is the most basic type of plot. It connects data points with lines.


```python
# Create simple data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Display the plot
plt.show()
```

## Adding Labels and Title

Good plots always have labels and titles to make them understandable.


```python
# Create data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create plot with labels and title
plt.plot(x, y)
plt.xlabel('X Axis Label')  # X-axis label
plt.ylabel('Y Axis Label')   # Y-axis label
plt.title('My First Plot')   # Plot title

# Display the plot
plt.show()
```

## Working with NumPy Arrays

Matplotlib works seamlessly with NumPy arrays, which is perfect since you already know NumPy!


```python
# Create data using NumPy
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine wave

# Create plot
plt.plot(x, y)
plt.xlabel('X values')
plt.ylabel('sin(X)')
plt.title('Sine Wave')
plt.grid(True)  # Add grid for better readability

plt.show()
```

## Multiple Lines on One Plot

You can plot multiple lines on the same plot to compare data.


```python
# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot multiple lines
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

# Add labels, title, and legend
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Sine and Cosine Waves')
plt.legend()  # Show legend to identify lines
plt.grid(True)

plt.show()
```

## Customizing Line Styles and Colors

You can customize how your lines look using colors, line styles, and markers.


```python
# Create data
x = np.linspace(0, 10, 50)

# Different line styles and colors
plt.plot(x, np.sin(x), 'b-', label='Solid blue', linewidth=2)
plt.plot(x, np.cos(x), 'r--', label='Dashed red', linewidth=2)
plt.plot(x, np.sin(x) * 0.5, 'g:', label='Dotted green', linewidth=2)
plt.plot(x, np.cos(x) * 0.5, 'm-.', label='Dash-dot magenta', linewidth=2)

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Customized Line Styles')
plt.legend()
plt.grid(True, alpha=0.3)  # alpha controls transparency

plt.show()
```

## Scatter Plot

Scatter plots are used to visualize the relationship between two variables. Each point represents one observation.


```python
# Create sample data
np.random.seed(42)  # For reproducibility
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5  # Linear relationship with noise

# Create scatter plot
plt.scatter(x, y, alpha=0.6)  # alpha controls transparency
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot')
plt.grid(True, alpha=0.3)

plt.show()
```

## Bar Chart

Bar charts are used to compare categorical data or show counts.


```python
# Create sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

# Create bar chart
plt.bar(categories, values, color=['skyblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.grid(True, axis='y', alpha=0.3)  # Only horizontal grid lines

plt.show()
```

## Histogram

Histograms show the distribution of numerical data. They're essential for understanding data quality and patterns.


```python
# Create sample data (normal distribution)
np.random.seed(42)
data = np.random.normal(100, 15, 1000)  # Mean=100, Std=15, 1000 points

# Create histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram - Data Distribution')
plt.grid(True, axis='y', alpha=0.3)

plt.show()
```

## Working with Pandas DataFrames

Since you know Pandas, this is where Matplotlib becomes really powerful! Pandas DataFrames have built-in plotting methods.


```python
# Create a sample DataFrame
df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [100, 120, 140, 110, 160, 180],
    'Expenses': [80, 85, 90, 95, 100, 105]
})

print("DataFrame:")
print(df)
print()

# Plot using pandas (it uses matplotlib under the hood)
df.plot(x='Month', y='Sales', kind='line', marker='o', title='Sales Over Time')
plt.ylabel('Sales ($)')
plt.grid(True, alpha=0.3)
plt.show()
```

## Multiple Columns from DataFrame

You can plot multiple columns from a DataFrame on the same plot.


```python
# Plot multiple columns
df.plot(x='Month', y=['Sales', 'Expenses'], kind='line', marker='o', title='Sales vs Expenses')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Bar Chart from DataFrame

You can easily create bar charts from DataFrames.


```python
# Create bar chart from DataFrame
df.plot(x='Month', y='Sales', kind='bar', color='steelblue', title='Sales by Month')
plt.ylabel('Sales ($)')
plt.xlabel('Month')
plt.xticks(rotation=0)  # Keep month labels horizontal
plt.grid(True, axis='y', alpha=0.3)
plt.show()
```

## Creating Subplots

Subplots allow you to create multiple plots in one figure. This is very useful for comparing different views of your data.


```python
# Create figure with subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # figsize controls width and height

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Plot 1: Line plot
axes[0, 0].plot(x, y1, 'b-')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('sin(X)')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Line plot
axes[0, 1].plot(x, y2, 'r-')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('cos(X)')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Bar chart
axes[1, 0].bar(categories, values, color='green')
axes[1, 0].set_title('Bar Chart')
axes[1, 0].set_xlabel('Categories')
axes[1, 0].set_ylabel('Values')
axes[1, 0].grid(True, axis='y', alpha=0.3)

# Plot 4: Scatter plot
np.random.seed(42)
x_scatter = np.random.randn(50)
y_scatter = np.random.randn(50)
axes[1, 1].scatter(x_scatter, y_scatter, alpha=0.6)
axes[1, 1].set_title('Scatter Plot')
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('Y')
axes[1, 1].grid(True, alpha=0.3)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

## Saving Figures

You can save your plots to files in various formats (PNG, PDF, SVG, etc.). This is essential for reports and documentation.


```python
# Create a plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(x, y, 'b-', linewidth=2)
plt.xlabel('X values', fontsize=12)
plt.ylabel('sin(X)', fontsize=12)
plt.title('Sine Wave - Saved Plot', fontsize=14)
plt.grid(True, alpha=0.3)

# Save the figure
plt.savefig('sine_wave.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'sine_wave.png'")

# You can also save as PDF
# plt.savefig('sine_wave.pdf', bbox_inches='tight')

plt.show()
```

## Real-World Example: Data Engineering Use Case

Let's create a practical example that a data engineer might encounter - visualizing data pipeline metrics.


```python
# Simulate ETL pipeline metrics over time
dates = pd.date_range('2024-01-01', periods=30, freq='D')
pipeline_data = pd.DataFrame({
    'Date': dates,
    'Records_Processed': np.random.randint(1000, 5000, 30),
    'Processing_Time_Seconds': np.random.uniform(10, 60, 30),
    'Success_Rate': np.random.uniform(0.95, 1.0, 30)
})

# Create subplots for different metrics
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Plot 1: Records processed over time
axes[0].plot(pipeline_data['Date'], pipeline_data['Records_Processed'], 
             marker='o', color='steelblue', linewidth=2)
axes[0].set_title('ETL Pipeline: Records Processed Over Time', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Records Processed')
axes[0].grid(True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# Plot 2: Processing time
axes[1].plot(pipeline_data['Date'], pipeline_data['Processing_Time_Seconds'], 
             marker='s', color='coral', linewidth=2)
axes[1].set_title('ETL Pipeline: Processing Time', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Processing Time (seconds)')
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)

# Plot 3: Success rate
axes[2].plot(pipeline_data['Date'], pipeline_data['Success_Rate'] * 100, 
             marker='^', color='green', linewidth=2)
axes[2].set_title('ETL Pipeline: Success Rate', fontsize=14, fontweight='bold')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Success Rate (%)')
axes[2].grid(True, alpha=0.3)
axes[2].tick_params(axis='x', rotation=45)
axes[2].set_ylim([90, 100])  # Set y-axis limits

plt.tight_layout()
plt.show()
```
