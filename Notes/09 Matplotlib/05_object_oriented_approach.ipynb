# 05 - Object-Oriented Approach

## Introduction

Matplotlib has two main interfaces: pyplot (which we've been using) and the object-oriented (OO) interface. The OO approach gives you more control and is better for complex plots and reusable code.

## What You'll Learn

- Understanding Figure and Axes objects
- OO approach vs pyplot
- Creating plots using OO interface
- Working with multiple axes
- When to use each approach


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Pyplot vs Object-Oriented Approach

**Pyplot (what we've used so far):**
- Simple, MATLAB-like interface
- Good for quick plots
- Uses global state
- Less control for complex plots

**Object-Oriented Approach:**
- More explicit and clear
- Better for complex plots
- More control over individual elements
- Easier to reuse and maintain
- Recommended for production code


```python
# OO Approach: Create Figure and Axes explicitly
fig, ax = plt.subplots(figsize=(10, 6))

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot using axes object
ax.plot(x, y, linewidth=2, label='sin(x)')
ax.set_xlabel('X values', fontsize=12)
ax.set_ylabel('Y values', fontsize=12)
ax.set_title('Object-Oriented Plot', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Working with Multiple Axes (OO Style)

The OO approach makes it easier to work with multiple subplots.


```python
# Create figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

x = np.linspace(0, 10, 100)

# Plot 1
axes[0, 0].plot(x, np.sin(x), 'b-', linewidth=2)
axes[0, 0].set_title('Sine Wave')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2
axes[0, 1].plot(x, np.cos(x), 'r-', linewidth=2)
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3
axes[1, 0].bar(['A', 'B', 'C', 'D'], [10, 20, 15, 25], color='green')
axes[1, 0].set_title('Bar Chart')
axes[1, 0].grid(True, axis='y', alpha=0.3)

# Plot 4
axes[1, 1].scatter(np.random.randn(50), np.random.randn(50), alpha=0.6)
axes[1, 1].set_title('Scatter Plot')
axes[1, 1].grid(True, alpha=0.3)

plt.suptitle('Multiple Plots - OO Approach', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
