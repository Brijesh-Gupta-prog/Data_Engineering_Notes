```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# List available styles
print("Available matplotlib styles:")
print(plt.style.available)
```

## Built-in Styles

Matplotlib comes with several built-in styles that change the overall appearance of your plots. Common styles include:
- `default`: The default matplotlib style
- `seaborn-v0_8`: Seaborn-like style (if seaborn is installed)
- `ggplot`: R's ggplot2 style
- `classic`: Matplotlib 1.x style
- `dark_background`: Dark theme
- `bmh`: Bayesian Methods for Hackers style


```python
# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Compare different styles
styles = ['default', 'ggplot', 'classic', 'dark_background']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for i, style in enumerate(styles):
    plt.style.use(style)
    row = i // 2
    col = i % 2
    axes[row, col].plot(x, y1, label='sin(x)', linewidth=2)
    axes[row, col].plot(x, y2, label='cos(x)', linewidth=2)
    axes[row, col].set_title(f'Style: {style}', fontweight='bold')
    axes[row, col].legend()
    axes[row, col].grid(True, alpha=0.3)

plt.suptitle('Comparison of Matplotlib Styles', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Reset to default
plt.style.use('default')
```

## Customizing Colors

Colors are crucial for effective visualization. You can use:
- Named colors: 'red', 'blue', 'green'
- Hex codes: '#FF5733', '#33FF57'
- RGB tuples: (0.1, 0.2, 0.5)
- Color maps: For gradients and heatmaps


```python
# Create data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

# Professional color palette
colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#6A994E']

plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
plt.xlabel('Categories', fontsize=12, fontweight='bold')
plt.ylabel('Values', fontsize=12, fontweight='bold')
plt.title('Bar Chart with Custom Color Palette', fontsize=14, fontweight='bold')
plt.grid(True, axis='y', alpha=0.3)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Font Customization

Customizing fonts improves readability and gives your plots a professional look.


```python
# Set global font properties
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12

# Create plot with custom fonts
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, linewidth=2, color='steelblue')
plt.xlabel('X Axis', fontweight='bold')
plt.ylabel('Y Axis', fontweight='bold')
plt.title('Plot with Custom Font Settings', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Reset to defaults
plt.rcParams.update(plt.rcParamsDefault)
```

## Professional Figure Aesthetics

Creating publication-quality figures requires attention to detail: spacing, layout, and overall appearance.


```python
# Create professional-looking plot
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = np.random.randint(50000, 100000, 12)

# Create figure with proper sizing
fig, ax = plt.subplots(figsize=(12, 6))

# Plot with professional styling
ax.plot(dates, sales, marker='o', linewidth=2.5, markersize=8, 
        color='#2E86AB', markerfacecolor='white', markeredgewidth=2, 
        markeredgecolor='#2E86AB', label='Monthly Sales')

# Customize axes
ax.set_xlabel('Month', fontsize=14, fontweight='bold')
ax.set_ylabel('Sales ($)', fontsize=14, fontweight='bold')
ax.set_title('Monthly Sales Report - 2024', fontsize=16, fontweight='bold', pad=20)

# Grid styling
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)

# Customize ticks
ax.tick_params(axis='both', which='major', labelsize=11)
plt.xticks(rotation=45, ha='right')

# Add value annotations
for i, (date, sale) in enumerate(zip(dates, sales)):
    if i % 2 == 0:  # Annotate every other point
        ax.annotate(f'${sale:,}', (date, sale), 
                   textcoords="offset points", xytext=(0,10), ha='center',
                   fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```
