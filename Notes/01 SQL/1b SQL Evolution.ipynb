**This is why RDBMS had to exist.**

### Core Problems

- Same data copied in multiple files

- "Employee" and "Department" had no enforced link

## 4. The Breakthrough Idea: Relational Model (1970)

**This is the turning point.**

### Edgar F. Codd (IBM, 1970)

**Proposed:**
- Data as tables (relations)
- Rows = records
- Columns = attributes
- Relationships via keys

**Radical ideas at the time:**
- Separate data from application logic
- Use mathematics (set theory) to query data
- Users describe what they want, not how to get it

## 5. Why SQL Was Created

SQL wasn't magic. It was a tool to talk to relational databases.

### What SQL Means
- **Structured Query Language**
- Declarative language

### Declarative vs Procedural

- **Procedural:** How to loop, how to fetch
- **Declarative:** What result you want

### Example:

```sql
SELECT department, AVG(salary)
FROM employees
GROUP BY department;
```

**You never explain:**
- Which index to use
- How to scan rows
- How to aggregate

**The database figures it out.**

**Drive this home:**

> "SQL shifted complexity from developers to the database engine."

## 7. Understanding ACID Properties

**ACID** is what makes databases reliable. Think of it as a guarantee that your data operations are safe and predictable.

### The Real-World Analogy: Bank Transfer

Imagine transferring $100 from your account to a friend's account. This involves two steps:
1. Deduct $100 from your account
2. Add $100 to your friend's account

**Without ACID:** What if step 1 succeeds but step 2 fails? Your money disappears! ðŸ’¸

**With ACID:** Either both steps succeed, or both fail. No money disappears. âœ…

### A - Atomicity: "All or Nothing"

**Simple definition:** A transaction is like a single, indivisible unit. Either everything happens, or nothing happens.

**Think of it as:** A light switch. It's either ON or OFF. There's no "halfway on."

**Example:**
```sql
BEGIN TRANSACTION;
  UPDATE accounts SET balance = balance - 100 WHERE id = 1;  -- Your account
  UPDATE accounts SET balance = balance + 100 WHERE id = 2;  -- Friend's account
COMMIT;
```

**What happens:**
- âœ… If both updates succeed â†’ Transaction commits. Money transferred.
- âŒ If either update fails â†’ Transaction rolls back. Both accounts unchanged.

**Key point:** You'll never see a state where money is deducted but not added, or vice versa.

---

### C - Consistency: "Rules Always Apply"

**Simple definition:** The database always follows its rules. Invalid states are impossible.

**Think of it as:** A vending machine. It won't give you a product if you don't pay enough. The rules are enforced.

**Example:**
```sql
-- Rule: Account balance cannot be negative
UPDATE accounts SET balance = balance - 500 WHERE id = 1 AND balance >= 500;
```

**What happens:**
- âœ… If balance is $600 â†’ Transaction succeeds. New balance: $100.
- âŒ If balance is $300 â†’ Transaction fails. Balance stays $300. Rule enforced!

**Key point:** Constraints (like "balance >= 0") are always maintained, even during transactions.

---

### I - Isolation: "Your Transaction Doesn't See Others' Mess"

**Simple definition:** Concurrent transactions don't interfere with each other. Each transaction sees a consistent snapshot of data.

**Think of it as:** Multiple people editing different parts of a document. They don't see each other's unsaved changes.

**Example Scenario:**
- **Transaction A:** Reading account balance (sees $1000)
- **Transaction B:** At the same time, updating balance to $1500
- **Transaction A:** Still sees $1000 until Transaction B commits

**What happens:**
- Without isolation: Transaction A might see partial updates (inconsistent data)
- With isolation: Transaction A sees a consistent snapshot, as if Transaction B didn't exist yet

**Key point:** Each transaction operates as if it's the only one running, even when many are running simultaneously.

---

### D - Durability: "Once Saved, It's Permanent"

**Simple definition:** Once a transaction commits, the changes are permanent, even if the system crashes.

**Think of it as:** Writing in permanent ink. Even if you spill coffee on the paper, the writing remains.

**Example:**
```sql
BEGIN TRANSACTION;
  INSERT INTO orders (customer_id, amount) VALUES (123, 500);
COMMIT;  -- Transaction is now durable
```

**What happens:**
- âœ… After COMMIT â†’ Data is written to disk
- ðŸ’¥ System crashes 1 second later â†’ Data is still there when system restarts
- âœ… Your order is not lost

**Key point:** Committed data survives power failures, crashes, and restarts. It's written to persistent storage.
