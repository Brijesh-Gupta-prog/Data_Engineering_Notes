# 06 - Loops

## Introduction

Loops allow you to repeat code multiple times. This is essential for processing lists of data or repeating actions.

## What You'll Learn

- `for` loops - iterate through sequences
- `while` loops - repeat while a condition is true
- `range()` function
- `break` and `continue` statements


## For Loop with range()

The `for` loop is used to iterate through a sequence. The `range()` function generates a sequence of numbers.

- `range(1, 6)` generates numbers from 1 to 5 (stops before 6)
- The code inside the loop is indented


```python
# For loop - iterate through a range
print("Counting from 1 to 5:")
for i in range(1, 6):  # range(1, 6) gives numbers 1, 2, 3, 4, 5
    print(i)
```

## For Loop with Lists

You can also loop through items in a list directly.


```python
# For loop - iterate through a list
print("Days of the week:")
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday","sunday"]
for day in days:
    print(day)
```

## While Loop

A `while` loop repeats code as long as a condition is `True`. Be careful to avoid infinite loops!


```python
# While loop - repeat while condition is true
print("Counting down:")
count = 5
while count > 0:
    print(count)
    count = count - 1  # Decrease count by 1
print("Blast off!")
```

## Break Statement

The `break` statement exits the loop immediately, even if the condition is still true.


```python
# Loop with break (exit early)
print("Finding first even number:")
for num in range(1, 10):
    if num % 2 == 0:  # Check if number is even
        print("Found:", num)
        break  # Exit the loop
```

## Continue Statement

The `continue` statement skips the rest of the current iteration and moves to the next one.


```python
# Loop with continue (skip iteration)
print("Printing odd numbers:")
for num in range(1, 10):
    if num % 2 == 0:  # Skip even numbers
        continue
    print(num)
```
