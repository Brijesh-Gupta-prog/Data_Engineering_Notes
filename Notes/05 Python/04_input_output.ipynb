## Introduction

Interactive programs need to get input from users. Python's `input()` function allows you to get text from the user.

## What You'll Learn

- How to get input from users using `input()`
- How to convert input to different data types
- Building interactive programs


## Getting Text Input

The `input()` function displays a prompt and waits for the user to type something. It always returns a string.


```python
# Get input from user
name = input("What is your name? ")
print("Nice to meet you,", name + "!")
```

## Converting Input to Numbers

Since `input()` always returns a string, you need to convert it to a number if you want to do math with it.

- Use `int()` to convert to an integer (whole number)
- Use `float()` to convert to a float (decimal number)


```python
# Get a number (input always returns a string, so convert it)
age = input("How old are you? ")
age = int(age)  # Convert string to integer
print("You are", age, "years old")
print("Next year you will be", age + 1)
```

```python
# Get a decimal number
height = float(input("What is your height in meters? "))
print("Your height is", height, "meters")
```
