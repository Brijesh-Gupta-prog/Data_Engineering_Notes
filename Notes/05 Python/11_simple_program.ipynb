## Step 1: Define Functions

First, we'll create functions to calculate BMI and determine the category.


```python
def calculate_bmi(weight, height):
    """Calculate Body Mass Index"""
    bmi = weight / (height ** 2)
    return bmi

def get_bmi_category(bmi):
    """Get BMI category based on BMI value"""
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"
```

## Step 2: Main Program

Now we'll create the main program that gets user input and displays results.


```python
# Main program
print("=== BMI Calculator ===")
print()

# Get user input
name = input("Enter your name: ")
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

# Calculate BMI
bmi = calculate_bmi(weight, height)
category = get_bmi_category(bmi)

# Display results
print()
print(f"Hello, {name}!")
print(f"Your BMI is: {bmi:.2f}")
print(f"Category: {category}")
```
