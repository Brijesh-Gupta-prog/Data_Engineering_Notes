# 09 - Functions

## Introduction

Functions are reusable blocks of code that perform a specific task. They help organize code and avoid repetition.

## What You'll Learn

- Defining functions with `def`
- Function parameters
- Return values
- Default parameters
- Multiple return values


## Simple Function

A function is defined using the `def` keyword followed by the function name and parentheses. The code inside is indented.


```python
# Simple function
def greet():
    print("Hello!")

# Call the function
greet()
```

## Function with Parameters

Functions can accept parameters (inputs) that you pass when calling the function.


```python
# Function with parameters
def greet_person(name):
    print("Hello,", name + "!")

greet_person("Alice")
greet_person("Bob")
```

## Function with Return Value

Functions can return values using the `return` statement. This allows you to use the result of the function.


```python
# Function that returns a value
def add_numbers(a, b):
    result = a + b
    return result

sum_result = add_numbers(5, 3)
print("Sum:", sum_result)
```

## Function with Default Parameters

You can provide default values for parameters. If the caller doesn't provide a value, the default is used.


```python
# Function with default parameter
def introduce(name, age=18):
    print("My name is", name, "and I am", age, "years old")

introduce("Charlie")  # Uses default age
introduce("Diana", 25)  # Uses provided age
```

## Multiple Return Values

Functions can return multiple values, which are returned as a tuple.


```python
# Function with multiple return values
def get_name_and_age():
    name = "Eve"
    age = 22
    return name, age

name, age = get_name_and_age()
print("Name:", name, "Age:", age)
```
