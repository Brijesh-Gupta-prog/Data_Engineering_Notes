## Print Statements

The `print()` function displays text or values on the screen. It's one of the most commonly used functions in Python.


```python
# Print statement - displays text on the screen
print("Hello, World!")
print("Welcome to Python Programming!")
```

## Printing Numbers

You can also print numbers directly without quotes.


```python
# You can print numbers
print(42)
print(3.14)
```

## Printing Multiple Items

You can print multiple items separated by commas. Python will automatically add spaces between them.


```python
# You can print multiple things
print("My name is", "Python")
print("I am", 30, "years old")
```
