# 07 - Lists

## Introduction

Lists are collections of items stored in a specific order. They are one of the most useful data structures in Python.

## What You'll Learn

- Creating lists
- Accessing list elements
- Adding and removing items
- List methods and operations
- List slicing


## Creating Lists

Lists are created using square brackets `[]` with items separated by commas.


```python
# Create a list
fruits = ["apple", "banana", "orange"]
print("Fruits:", fruits)
```

## Accessing List Elements

List indices start at 0 (not 1!). You can use:
- Positive indices: `[0]` for first item, `[1]` for second, etc.
- Negative indices: `[-1]` for last item, `[-2]` for second-to-last, etc.


```python
# Access list elements (index starts at 0)
print("First fruit:", fruits[0])
print("Second fruit:", fruits[1])
print("Last fruit:", fruits[-1])  # -1 means last item
```

## Adding and Removing Items

Lists are mutable, meaning you can change them after creation.


```python
# Add items to list
fruits.append("grape")
print("After adding grape:", fruits)

# Remove items
fruits.remove("banana")
print("After removing banana:", fruits)
```

## List Length and Useful Functions

The `len()` function returns the number of items in a list. There are also built-in functions for numeric lists.


```python
# List length
print("Number of fruits:", len(fruits))

# List with numbers
numbers = [1, 2, 3, 4, 5]
print("\nNumbers:", numbers)
print("Sum:", sum(numbers))
print("Maximum:", max(numbers))
print("Minimum:", min(numbers))
```

## Looping Through Lists

You can use a `for` loop to process each item in a list.


```python
# Loop through list
print("All fruits:")
for fruit in fruits:
    print("-", fruit)
```

## List Slicing

Slicing allows you to get a portion of a list using the syntax `[start:end]`. The end index is not included.


```python
# List slicing (get part of list)
fruits = ["apple", "banana", "orange", "grape", "mango"]
print("All fruits:", fruits)
print("First two fruits:", fruits[0:2])
print("Last two fruits:", fruits[-2:])
```
