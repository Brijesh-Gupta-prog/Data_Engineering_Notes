# 10 - String Operations

## Introduction

Strings are sequences of characters. Python provides many built-in methods to work with strings.

## What You'll Learn

- String methods (upper, lower, replace, split, etc.)
- String formatting
- String slicing
- Checking if strings contain substrings


## Basic String Operations

Let's start with some fundamental string operations.


```python
# Basic string operations
text = "Hello World"
hello = "Hello"

# Length of string
print("Length:", len(text))

# Convert to uppercase
print("Uppercase:", text.upper())

# Convert to lowercase
print("Lowercase:", text.lower())

# Capitalize first letter
print("Capitalize:", text.capitalize())
```

## String Methods: Replace and Split

- `replace()` - replaces one substring with another
- `split()` - splits a string into a list of words


```python
# Replace text
new_text = text.replace("World", "Python")
print("After replace:", new_text)

# Split string into list
words = text.split()
print("Words:", words)
```

## Joining Strings

The `join()` method combines a list of strings into a single string.


```python
# Join strings
fruits = ["apple", "banana", "orange"]
joined = ", ".join(fruits)
print("Joined:", joined)
```

## Checking if String Contains Substring

Use the `in` keyword to check if a substring exists in a string.


```python
# Check if string contains substring
if "Hello" in text:
    print("'Hello' is in the text")
```

## String Formatting

There are multiple ways to format strings in Python. F-strings (formatted string literals) are the modern and recommended way.


```python
name = "Alice"
age = 25


# String formatting (old way)
message = "My name is %s and I am %d years old" % (name, age)
print(message)

# String formatting (new way - f-strings) - RECOMMENDED
message = f"My name is {name} and I am {age} years old"
print(message)
```

## String Slicing

You can extract parts of a string using slicing, similar to list slicing.


```python
# String slicing
text = "Python Programming"
print("First 6 characters:", text[0:6])
print("Last 5 characters:", text[-5:])
```
