## Introduction

Python supports various operations for working with numbers and strings. This notebook covers arithmetic and string operations.

## What You'll Learn

- String operations (concatenation, repetition)
- Type conversion


## Arithmetic Operations

Python supports all basic mathematical operations:
- `+` Addition
- `-` Subtraction
- `*` Multiplication
- `/` Division (returns float)
- `//` Floor division (returns integer)
- `%` Modulus (remainder)
- `**` Exponentiation (power)


```python
# Arithmetic Operations
a = 10
b = 3

print("Addition:", a + b)        # 13
print("Subtraction:", a - b)      # 7
print("Multiplication:", a * b)   # 30
print("Division:", a / b)         # 3.333...
print("Floor Division:", a // b)  # 3 (whole number division)
print("Modulus:", a % b)          # 1 (remainder)
print("Exponent:", a ** b)        # 1000 (10 to the power of 3)
```

## String Operations

Strings can be combined (concatenated) using the `+` operator, and repeated using the `*` operator.


```python
# String Operations
first_name = "John"
last_name = "Doe"

full_name = first_name + " " + last_name  # Concatenation
print("Full name:", full_name)

# Repeat a string
print("Ha" * 3)  # Prints: HaHaHa
```

## Type Conversion

When combining numbers and strings, you need to convert the number to a string first using `str()`.


```python
# Combining numbers and strings (need to convert)
age = 25
message = "I am " + str(age) + " years old"
print(message)
```
