````python
# Exercise 1: Verification Helper
# Run this cell to check your setup

import subprocess
import os

def run_cmd(command):
    try:
        result = subprocess.run(command, shell=True, capture_output=True, text=True, check=True)
        return result.stdout.strip()
    except:
        return "Error running command"

print("=== Exercise 1 Verification ===")
print("\n1. Check if in a git repository:")
try:
    result = run_cmd("git rev-parse --git-dir")
    print(f"âœ… Git repository found: {result}")
except:
    print("âŒ Not in a git repository. Run 'git init' first.")

print("\n2. Check git status:")
status = run_cmd("git status --short")
if not status or "nothing to commit" in run_cmd("git status"):
    print("âœ… Working directory is clean")
else:
    print(f"âš ï¸  Uncommitted changes: {status}")

print("\n3. Check commit history:")
log = run_cmd("git log --oneline -1")
if log:
    print(f"âœ… Latest commit: {log}")
else:
    print("âŒ No commits found")
````

## Exercise 13: Undoing Changes

### Objective
Learn to undo changes at different stages.

### Steps

1. **Create a test file and commit it**:
   ```bash
   echo "# Test file" > test_undo.txt
   git add test_undo.txt
   git commit -m "Add test file for undo exercise"
   ```

2. **Make a change and unstage it**:
   ```bash
   echo "# Modified" >> test_undo.txt
   git add test_undo.txt
   git reset HEAD test_undo.txt  # Unstage
   git status
   ```

3. **Discard working directory changes**:
   ```bash
   git checkout -- test_undo.txt
   cat test_undo.txt  # Should be back to original
   ```

4. **Make a commit you want to undo** (keep changes):
   ```bash
   echo "# New change" >> test_undo.txt
   git add test_undo.txt
   git commit -m "Change to undo"
   git reset --soft HEAD~1
   git status  # Changes still staged
   ```

5. **Make a commit and undo it completely**:
   ```bash
   git commit -m "Change to undo completely"
   git reset --hard HEAD~1
   git status  # Changes gone
   ```

6. **Practice with git revert** (safer for shared commits):
   ```bash
   echo "# Safe change" >> test_undo.txt
   git add test_undo.txt
   git commit -m "Safe change"
   git log --oneline -1
   git revert HEAD  # Creates new commit undoing the change
   git log --oneline -2
   ```

### âš ï¸ Important Notes
- `git reset --hard` is destructive - use carefully
- `git revert` is safer for commits already pushed
- `git checkout -- file` discards uncommitted changes

### Verification
```bash
git log --oneline -5
git status
```

### âœ… Checkpoint
Understand different ways to undo changes!

## Exercise 16: Troubleshooting Common Issues

### Objective
Practice resolving common Git problems.

### Scenario 1: "Your branch is behind 'origin/main'"

**Steps:**
```bash
# Fetch latest changes
git fetch origin

# See what's different
git log HEAD..origin/main

# Update your branch
git pull origin main
# OR
git rebase origin/main
```

### Scenario 2: Accidentally committed to wrong branch

**Steps:**
```bash
# Undo commit but keep changes
git reset --soft HEAD~1

# Switch to correct branch
git checkout correct-branch

# Commit again
git commit -m "Correct commit message"
```

### Scenario 3: Want to change last commit message

**Steps:**
```bash
git commit --amend -m "New commit message"
```

### Scenario 4: Added file to commit that shouldn't be there

**Steps:**
```bash
# Unstage the file
git reset HEAD unwanted_file.txt

# Amend the commit
git commit --amend --no-edit
```

### Scenario 5: Need to undo last commit but keep working

**Steps:**
```bash
# Undo commit, keep changes staged
git reset --soft HEAD~1

# Or undo commit and unstage, keep changes
```
